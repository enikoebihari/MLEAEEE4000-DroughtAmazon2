# Hansen forest mask preprocessing

## Preprocesses Hansen continuous forest mask from 2000-2024 

(Hansen data set generated using GEE javascript API https://code.earthengine.google.com/33b627316962cec86c281588aee29cfa)
- rescales and reprojects to match the LAI data set

## Setup

```python
# import packages
import xarray as xr
import rioxarray as rio
import numpy as np
import pandas as pd
import os
import glob
import rasterio
import re
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
from rasterio.enums import Resampling 

# load in data
# --------------------------------------------------------------------------------------------
# --------------------------------------------------------------------------------------------

# MODIS LAI 
# --------------------------

# set the path and open the data set
lai_path = "/home/jovyan/MLEAEEE4000-DroughtAmazon2/data/modisLAI/MODIS_LAI_monthly.nc"
lai_xr = xr.open_dataset(lai_path)

# Set spatial dimensions and CRS
lai_xr = lai_xr.rio.set_spatial_dims(x_dim="lon", 
                                     y_dim="lat", 
                                     inplace=False)
lai_xr = lai_xr.rio.write_crs("EPSG:4326", 
                              inplace=True) 

# print some stuff out to check
# print("LAI:", lai_xr) #full data set
# print("CRS:", lai_xr['lai'].rio.crs) #crs
# print("dimensions:", lai_xr['lai'].dims) #dimensions
# print("time steps:", lai_xr.time.values)  # dates

# plot
lai_xr.lai.isel(time=210).plot(figsize=(8,6))

# Hansen mask
# --------------------------

# set the path and open the data set
# preprocessed using https://code.earthengine.google.com/33b627316962cec86c281588aee29cfa
hansenMask_tif = rio.open_rasterio("/home/jovyan/MLEAEEE4000-DroughtAmazon2/data/hansenMask_mode_27m_v2.tif")

# print some stuff out to check
print("Hansen mask:", hansenMask_tif) # CRS
print("CRS:", hansenMask_tif.rio.crs) # CRS
print("Affine transform:", hansenMask_tif.rio.transform()) # Affine transform
print("Pixel size:", hansenMask_tif.rio.resolution()) # Pixel size 
print("Bounds:", hansenMask_tif.rio.bounds()) # Bounds (xmin, ymin, xmax, ymax)
print("Width (px):", hansenMask_tif.rio.width) # pixel width
print("Height (px):", hansenMask_tif.rio.height) # pixel height

# plot
# ran into memory issues
```

## Reproject

```python
# reproject Hansen forest mask to match the pixel size and crs of the LAI data set

# set target spatial parameters
target_crs = lai_xr.rio.crs #crs
target_transform = lai_xr.rio.transform() #transformation
target_shape = lai_xr.rio.shape # spatial dimensions

print('target CRS:',target_crs)
print('target transformation:',target_transform)
print('target shape:',target_shape)

# reproject the dataset
hansenMask_reproject_tif = hansenMask_tif.rio.reproject(
    dst_crs=target_crs,
    transform=target_transform,
    shape=target_shape,
    resampling=Resampling.nearest   
)

# visualize
hansenMask_reproject_tif.squeeze().plot(figsize=(8, 6))
plt.title("Reprojected Hansen")
plt.show()
```

## Export

```python
# export
hansenMask_reproject_tif.rio.to_raster("/home/jovyan/MLEAEEE4000-DroughtAmazon2/data/hansenMask_mode_50m_v2.tif")
```
