This script creates a GeoJSON export of the Madre de Dios political boundaries. It uses the GAUL dataset in the Google Earth Engine API, filters for the Madre de Dios political boundaries, and exports as a GeoJSON to the local directory.

```python
# import GEE
import ee
# Trigger the authentication flow
ee.Authenticate(force=True)
```

```python
# initialize the project library
ee.Initialize(project='columbia-phd')
```

```python
import rasterio
import geemap
```

Getting the Madre de Dios feature from the ee feature collection, to be used for clipping rasters to political extent. 

```python
# get the aoi boundaries by filtering the GAUL dataset
gaul = ee.FeatureCollection("FAO/GAUL/2015/level1") # get the FAO feature collection
peru = gaul.filter(ee.Filter.eq('ADM0_NAME', 'Peru')) # Filter Peru first (ADM0_NAME = country)
aoi = peru.filter(ee.Filter.eq('ADM1_NAME', 'Madre de Dios')) # Filter Madre de Dios region (ADM1_NAME = admin1 region)
```

```python
import json
```

```python
# Convert to GeoJSON dictionary
geojson_dict = aoi.getInfo()

# Save to file
with open("/home/jovyan/MLEAEEE4000-DroughtAmazon2/data/MDD_political_boundary.geojson", "w") as f:
    json.dump(geojson_dict, f)
```
